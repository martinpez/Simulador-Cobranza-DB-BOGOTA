````python
// Definir las funciones correctamente
function Cursor_Notallow(){
    const disabledInputs = document.querySelectorAll(
        'input[disabled], input[aria-disabled="true"], input[readonly]'
    );
    disabledInputs.forEach(input => {
        const widget = input.closest('.k-widget');
        if (widget) {
            widget.style.cursor = 'not-allowed';
            widget.style.color = '#000';
        }

        const wrap = input.closest('.k-numeric-wrap, .k-dropdown-wrap, .k-picker-wrap, .k-multiselect');
        if (wrap) { 
            wrap.style.cursor = 'not-allowed'; 
            widget.style.color= '#000'
        }
    });
}


function isInsideToggleCard(element) {
    var card = element.closest('.card1');
    return card && card.querySelector('.toggle-switch');
}

function Color_Inputs(){
    var inputs = document.querySelectorAll("input, textarea");
    for (var i = 0; i < inputs.length; i++) {
            var input = inputs[i];
            var isDisabled = input.disabled === true || input.readOnly === true || window.getComputedStyle(input).pointerEvents === "none";
        
            // EXCEPCIÓN para la card de Consolidación 
            if (isInsideToggleCard(input)) {
            continue;
        }
        
        if (isDisabled) {
            input.style.backgroundColor = "#d9d9d9";
            input.style.color = "#6b6b6b";
            if (input.parentElement) {
                input.parentElement.style.backgroundColor = "#d9d9d9";
            }
        }
    }
}

setTimeout(function () {
    Cursor_Notallow();
    Color_Inputs();
}, 7000);
````

````python
function syncObservationsWithSox() {

  const OBSERVATION_SOX_MAP = [
    { obs: 'be70a202-71a9-40ea-851b-945702693b51', sox: 'f3979225-f563-48a2-a206-6b5866a7dc6c' },
    { obs: '637cda5e-a8da-499a-98be-564521dd6c25', sox: '07b4e087-95c8-4867-b91f-1f9e9a4a1ea0' },
    { obs: '96c93177-4705-4bd2-ac50-e304c007afa3', sox: 'b24357e4-d1be-443d-8fa0-5b8790a1c508' },
    { obs: '24e68f6c-b401-40d9-bb2d-ec6d246426f9', sox: 'd4f89a7c-0207-4756-9bd7-e2e669ac3ce0' },
    { obs: '68d8ce24-c9fd-440b-995a-7ff027f628b6', sox: 'eec3136d-46bf-438c-b7cc-4aaa5fba776b' }
  ];

  const MAX_SOX_LENGTH = 1100;

  OBSERVATION_SOX_MAP.forEach(({ sox, obs }) => {

    const soxField = document.getElementById(sox);
    const obsField = document.getElementById(obs);
    if (!soxField || !obsField) return;

    const counter = soxField.parentElement.querySelector('.sox-counter');
    if (!counter) return;

    const soxValue = (getFieldValue(sox) || '').toUpperCase();
    const length = soxValue.length;

    // inicializar último valor válido
    if (obsField.dataset.lastValue === undefined) {
      obsField.dataset.lastValue = obsField.value || '';
    }

    // contador
    counter.innerText = `${length} / ${MAX_SOX_LENGTH}`;

    if (length >= MAX_SOX_LENGTH) {

      counter.style.color = '#dc3545';
      counter.style.fontWeight = '600';

      obsField.style.border = '1px solid #dc3545';

      // listener UNA sola vez
      if (!obsField.dataset.locked) {
        obsField.addEventListener('input', (event) => {

          const currentSox = (getFieldValue(sox) || '').toUpperCase();
          if (currentSox.length < MAX_SOX_LENGTH) return;

          // bloquear solo inserciones
          if (
            event.inputType === 'insertText' ||
            event.inputType === 'insertFromPaste'
          ) {
            obsField.value = obsField.dataset.lastValue;
            toastr.warning(
              'Has alcanzado el límite máximo de caracteres permitido por el SOX.'
            );
          } else {
            //  permitir borrar
            obsField.dataset.lastValue = obsField.value;
          }
        });

        obsField.dataset.locked = 'true';
      }

    } else {
      //  estado normal
      obsField.dataset.lastValue = obsField.value;
      obsField.style.border = '';

      counter.style.color = '#6c757d';
      counter.style.fontWeight = 'normal';
    }

  });

}
````

#  Sincronizacion de Observaciones hacia Campos SOX

## Informacion General

* **Ambiente:** Test (implementado) / Produccion (pendiente)

## Descripcion Funcional General
Se implemento una serie de eventos de tipo **General** para automatizar el reflejo de informacion entre los campos de "Observaciones" y los campos "SOX" de cada mecanismo. 

La logica garantiza que:
1. Se identifique un bloque delimitado especifico: `LLLOBSERVACIONESXX ... LLL`.
2. Se reemplace unicamente el contenido dentro de ese bloque.
3. Se mantenga intacta la informacion estructural y previa que ya exista en el campo SOX.

---

## Detalle de Eventos por Mecanismo

### 1. Mecanismo: Ampliacion
* **Nombre del evento:** Reflejar observacion en el campo de sox Ampliacion
* **Comportamiento:** Actualiza el campo SOX Ampliacion con el texto digitado en Observaciones, manteniendo la persistencia de la data previa fuera del delimitador.

### 2. Mecanismo: Consolidacion
* **Nombre del evento:** Reflejar Observaciones en sox Consolidacion
* **Comportamiento:** Actualiza exclusivamente el bloque de observaciones delimitado dentro del SOX Consolidacion, conservando la estructura existente.

### 3. Mecanismo: Cancelacion Total
* **Nombre del evento:** Reflejar campo observaciones en el campo sox Cancelacion
* **Comportamiento:** Inserta la informacion en el segmento correspondiente del SOX Cancelacion sin afectar el resto del contenido tecnico almacenado.

### 4. Mecanismo: Pago de Mora
* **Nombre del evento:** ValueChange Sox mora reflejar observaciones a Sox
* **Disparador:** Evento `ValueChange`.
* **Comportamiento:** Asegura la actualizacion controlada del texto delimitado en el SOX Mora ante cualquier cambio en el campo de origen.

### 5. Mecanismo: Novacion
* **Nombre del evento:** Reflejar Observaciones en campo sox Novacion
* **Comportamiento:** Reemplaza el bloque delimitado en el SOX Novacion preservando la informacion estructural previamente almacenada.

---

## Detalle Tecnico de la Logica
Para todos los casos, la implementacion utiliza el metodo de reemplazo de cadenas para localizar los delimitadores de control.

**Estructura del bloque afectado:**
`LLLOBSERVACIONESXX [Contenido de Observaciones] LLL`

**Flujo de ejecucion:**
1. El usuario digita en el campo de **Observaciones**.
2. El evento captura el valor del campo **SOX** correspondiente.
3. Se identifica la posicion de `LLLOBSERVACIONESXX` y el cierre `LLL`.
4. Se concatena la nueva cadena manteniendo los prefijos y sufijos originales del SOX.
5. Se actualiza el valor del componente SOX en la interfaz y el modelo.

#### Reflejar observacion en el campo de sox Ampliacion

## 3. Problemas Identificados

### Problema 1 — sessionStorage incompleto

Los valores `PorcAmpliacionIntCte` y `PorcAmpliacionIntMora` solo se guardaban en `sessionStorage` dentro del bloque condicional `if (EdadMoraCl === '1-30 Días')`. Para cualquier otra edad de mora esos valores quedaban `undefined`, haciendo que todas las funciones disparadas por `ValueChange` calcularan sobre cero.

### Problema 2 — Fórmula incorrecta en Abono con Dtos Max

Para rangos superiores a 30 días, el campo **Abono con Dtos Max** se calculaba como:

```javascript
//  INCORRECTO
AbonConDescMax = PrimaU + otrosCargos + 20000
```

Esta fórmula omitía la porción de interés corriente que el cliente **sí debe pagar** según el porcentaje de la edad de mora. Al ingresar ese valor en el Pago al SNR, el sistema interpretaba que el 100% del interés corriente quedaba pendiente, arrojando incorrectamente el **100%** en el campo `%Baja en cuenta Int Cte` en vez del **50%** esperado (ejemplo: cliente 61–120 días).

## 4. Solución Implementada

### 4.1 Corrección 1 — `CalculosAmpliacion()`

Se movió el bloque de almacenamiento en `sessionStorage` fuera del condicional de edad de mora, de modo que siempre se guarden los valores correctos independientemente del rango:

```javascript
//  ANTES — solo se guardaba para 1-30 días
if (e.dataItem.EdadMoraCl == "1-30 Días") {
    sessionStorage.PorcAmpliacionIntCte  = PorcAmpliacionIntCte;
    sessionStorage.PorcAmpliacionIntMora = PorcAmpliacionIntMora;
}

// DESPUÉS — se guarda para todas las edades
sessionStorage.EdadMoraCl           = e.dataItem.EdadMoraCl;
sessionStorage.PorcAmpliacionIntCte  = porcDescIntCteIcs;
sessionStorage.PorcAmpliacionIntMora = porcIntMora;
```

### 4.2 Corrección 2 — Fórmula de Abono con Dtos Max

Se unificó la fórmula en todas las funciones (`CalculosAmpliacion`, `InteresCorrienteAmpliacion`, `InteresesMoraIcs`, `OtrosCargosAmpliacion`, `PrimaUnicaAmpliacion`):

| Condición | Fórmula Anterior  | Fórmula Corregida  |
|---|---|---|
| 1–30 días (sin descuento) | `PrimaU + InteCte*(porc/100) + otrosCargos + 20.000` | `PrimaU + InteCte + IntMora + otrosCargos + 20.000` |
| > 30 días (con descuento) | `PrimaU + otrosCargos + 20.000` | `PrimaU + InteCte*(1 – porc/100) + otrosCargos + 20.000` |

```javascript
//  CORRECTO — para edades > 30 días
AbonConDescMax =
    parseInt(PrimaU      ?? 0) +
    Math.round(parseInt(InteCte ?? 0) * (1 - maxPorcDescIntCte / 100)) +
    parseInt(otrosCargos ?? 0) +
    20000;
```

**Interpretación de la fórmula para > 30 días:**

- **PrimaU**: se paga al 100% (sin descuento).
- **InteCte × (1 – porc/100)**: se paga solo la fracción no condonada (ej. 50% para 61–120 días).
- **Intereses en Mora**: se descuentan al 100% → el cliente no paga mora.
- **otrosCargos**: se pagan al 100%.
- **+ 20.000**: colchón fijo aplicado siempre al campo Abono con Dtos Max.

### 4.3 Corrección 3 — Comportamiento del campo %Baja en cuenta Int Cte

Se redefinió el comportamiento de `InteresCorrienteAmpliacion()` para eliminar el tope forzado en el porcentaje mostrado:

## 7. Flujo de Ejecución Corregido

### Al cargar el cliente (`CalculosAmpliacion`)

1. Se consulta la BD: `SELECT PorcAmpliacionIntCte, PorcAmpliacionIntMora WHERE RangoDias3 = EdadMoraCl`.
2. Se determina si aplica campaña (usa tasas de campaña) o no (usa tasas de la tabla).
3. Se guardan los porcentajes en `sessionStorage` para **todas** las edades de mora.
4. Se calculan los descuentos (`DescIntCte`, `DescIntMora`, `TotalDescuento`).
5. Se calcula `AbonConDescMax` con la fórmula corregida según edad de mora.
6. Se setean todos los campos de la pantalla.

### Al cambiar el Pago al SNR (`InteresCorrienteAmpliacion` + `InteresesMoraIcs`)

1. Se leen los valores actuales de todos los campos desde la interfaz.
2. Se calcula el % real de baja en interés corriente: `((PrimaU + InteCte + otrosCargos + 20.000 - pagoSNR) / InteCte)`.
3. Se muestra el % real en pantalla (sin tope) para informar al asesor.
4. Se recalcula `AbonConDescMax` con el tope máximo parametrizado (no cambia al ingresar el pago).
5. Si el pago ingresado es menor al Abono Mín. sin Dcto., se muestra un `toastr.warning`.
